```python
# Check data types
df.info()
```

```python
# Convert sales_date to datetime
df['sales_date'] = pd.to_datetime(df['sales_date'])
df.info()
```

```python
# Set the sale_date column as the index
df = df.set_index('sales_date')
df.head(5)
```

```python
# Resample data to monthly frequency and compute the sum
df_monthly = df.resample('M').sum()
df_monthly
```

```python
# Map the month number to short month name
df_monthly['month_name'] = df_monthly.index.strftime('%b')
df_monthly
```
